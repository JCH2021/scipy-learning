```python
import numpy as np
from scipy import linalg
A = np.array([[1,2],[2,1]])
print(A)
print('-'*20)
print(linalg.inv(A))
# A 的逆矩阵
print('-'*20)
print(A.dot(linalg.inv(A)))
# A 乘 A逆，数据复杂的话，可能对角线外不为0，那是不确定度，近似于0
```

用矩阵的方式解多元方程

$$x+y+z=0$$
$$x=1$$
$$y=-1$$

系数矩阵为：

$$\left[\begin{matrix}1 & 1 & 1 \cr 1 & 0 & 0 \cr 0 & 1 & 0\end{matrix}\right]$$ 

值矩阵为：

$$\left[\begin{matrix}0 \cr 1 \cr -1\end{matrix}\right]$$ 

计算公式为：

$$\left[\begin{matrix}x \cr y \cr z\end{matrix}\right]=\left[\begin{matrix}1 & 1 & 1 \cr 1 & 0 & 0 \cr 0 & 1 & 0\end{matrix}\right]^{-1} \left[\begin{matrix}0 \cr 1 \cr -1\end{matrix}\right]=\left[\begin{matrix}1 \cr -1 \cr 0\end{matrix}\right]$$ 

```python
import numpy as np
from scipy import linalg
C = np.array([[1,1,1],[1,0,0],[0,1,0]])
# A 的逆矩阵
R = np.array([[0],[1],[-1]])
np.linalg.solve(C, R)
```

```python
A = np.array([[1,2],[3,4]])
linalg.det(A)
#计算行列式
```

```python
A = np.array([[1, 2], [3, 4]])
la, v = linalg.eig(A)
l1, l2 = la
#计算矩阵的特征值
print(l1, l2)
v1, v2 = v
print(v1,v2)
#计算矩阵的特征向量
```
