对贝塞尔函数的一重定积分
$$I=\int_{0}^{3} J_{2}(x)+1\,dx$$

```python
import scipy.integrate as integrate
import scipy.special as special
result = integrate.quad(lambda x: special.jv(2,x)+1, 0, 3)
#函数用来计算定积分，参数一是表达式，参数一可自行定义，可以对积分嵌套
#参数二三分别是下上限
# jv 计算贝塞尔函数的值，参数一是阶
result
#结果一是积分结果，二是对结果绝对误差的估计
```

对函数的多重定积分
$$I=\int_{y=0}^{6}\int_{x=0}^{x=y} x \,dxdy=36$$

```python
from scipy.integrate import dblquad
area = dblquad(lambda x, y: x, 0, 6, lambda x: 0, lambda y: y)
#第一个函数就是外层积分和下上限
#函数二就是内部下限，函数三是内部上限
# x= lambda y:y = y
area
```
